Applying various Image Filtering techniques of Sobel, Averaged, Median and Gaussian Filters

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_img(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    if img is None:
        print(f"Error: unable to load image {img_path}")
    return img

def plot_img(img, filtered_img, filter_name):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(filtered_img, cmap='gray')
    plt.title(filter_name)
    plt.axis('off')

    plt.show()

def sobel_filter(img):
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
    sobel = np.sqrt(sobelx**2 + sobely**2)
    return sobel

def median_filter(img):
    median = cv2.medianBlur(img, 5)
    return median

def gaussian_filter(img):
    gaussian = cv2.GaussianBlur(img, (5, 5), 0)
    return gaussian

def averaging_filter(img):
    averaging = cv2.blur(img, (5, 5))
    return averaging

def main():
    img_path = input("Enter path to image file: ").strip()  #strip() removes extra spaces before or after img path to avoid errors in img loading
    img = load_img(img_path)

    if img is not None:
        while True:
            print("\nChoose a filter to apply:")
            print("1. Sobel filter")
            print("2. Median filter")
            print("3. Gaussian filter")
            print("4. Averaging filter")
            print("5. Exit")

            try:
                ch = int(input("Enter choice (1-5): "))

                if ch == 1:
                    res = sobel_filter(img)
                    plot_img(img, res, 'Sobel Filter Output')
                
                elif ch == 2:
                    res = median_filter(img)
                    plot_img(img, res, 'Median Filter Output')
                
                elif ch == 3:
                    res = gaussian_filter(img)
                    plot_img(img, res, 'Gaussian Filter Output')
                
                elif ch == 4:
                    res = averaging_filter(img)
                    plot_img(img, res, 'Averaging Filter Output')
                
                elif ch == 5:
                    break
                
                else:
                    print("Invalid choice. Please select a number between 1 and 5.")
            
            except ValueError:
                print("Invalid input. Please enter a valid number.")
    else:
        print("Provide a valid image path.")

if __name__ == "__main__":
    main()
```

```python
# cv2: OpenCV library used for image processing.
# numpy: For numerical operations (e.g., creating arrays and applying mathematical operations).
# matplotlib.pyplot: For visualizing the images and results.


# load_img(img_path)
# Purpose: Loads the image from the provided path in grayscale mode.

# cv2 Functionality:
# cv2.imread(img_path, cv2.IMREAD_GRAYSCALE):
# Reads the image in grayscale mode (1 channel instead of 3 for RGB).
# If the image cannot be read (e.g., invalid path), it returns None.


# plot_img(img, filtered_img, filter_name)
# Purpose: Displays the original image alongside the filtered image.

# matplotlib Functionality:
# plt.subplot: Creates subplots to display multiple images side by side.
# plt.imshow: Displays an image in grayscale mode (cmap='gray').
# plt.title: Adds a title to the subplot.
# plt.axis('off'): Hides axis labels for cleaner visualization.
    

# sobel_filter(img)
# Purpose: Applies the Sobel filter to calculate gradients in both x and y directions and combines them.

# cv2 Functionality:
# cv2.Sobel(img, cv2.CV_64F, dx, dy, ksize=5):
# Computes the derivative in the x (dx=1) and y (dy=1) directions.
# cv2.CV_64F specifies that the resulting image or matrix will store each pixel as a 64-bit floating-point number.
# cv2.CV_64F: Ensures that the calculated gradients (sobelx and sobely) can include negative values and have high precision.
# ksize=5: Kernel size for the Sobel operator (5x5 matrix).
# np.sqrt: Combines the gradients using Euclidean distance (sqrt(sobelx^2 + sobely^2)).
    
# Sobel Filter:
# Detects edges by computing the gradient in horizontal and vertical directions.
# Highlights boundaries where pixel intensity changes rapidly.


# median_filter(img)
# Purpose: Applies a median filter to reduce noise.
# cv2 Functionality:
# cv2.medianBlur(img, 5):
# Uses a 5x5 kernel to replace each pixel with the median value of its neighboring pixels.

# Median Filter:
# Replaces pixel values with the median of neighboring pixels.
# Preserves edges while removing noise (e.g., random white or black dots).


# gaussian_filter(img)
# Purpose: Applies a Gaussian filter to smooth the image.
# cv2 Functionality:
# cv2.GaussianBlur(img, (5, 5), 0):
# Convolves the image with a Gaussian kernel of size 5x5.
# 0 indicates that the standard deviation of the kernel is computed automatically.

# Gaussian Filter:
# Smooths the image using a Gaussian kernel.
# Reduces noise more effectively than a simple average but can slightly blur edges.


# averaging_filter(img)
# Purpose: Applies an averaging (box) filter to smooth the image.
# cv2 Functionality:
# cv2.blur(img, (5, 5)):
# Replaces each pixel value with the average of its neighbors within a 5x5 kernel.

# Averaging Filter:
# Averages the pixel intensities over a kernel.
# Effective for noise reduction but blurs the image significantly.


# Why Kernel Size = 5 in Your Code?
# Kernel Size (Odd Numbers): Kernels are typically odd-sized (e.g., 3×3, 5×5, 7×7) to ensure a well-defined center pixel.

# Balance: A 5×5 kernel strikes a balance between capturing local details and reducing noise or smoothing the image.
# Image Resolution: Larger kernel sizes are better for higher-resolution images, while smaller kernel sizes suit lower-resolution images.
# Moderate Noise: The code assumes the image might have moderate noise and hence uses 
# 5×5 for filters like Median, Gaussian, and Averaging.

# When to Choose Other Kernel Sizes?
# Smaller Kernels (3×3):
# If the image has fine details or edges that you want to preserve.
# Use when noise levels are low.
# Larger Kernels (7×7 or more):
# If the image is heavily noisy or requires significant smoothing.
# Use when detail preservation is less important than noise removal.



# Questions
# What is the role of the ksize parameter in the Sobel filter?
# Answer: The ksize parameter specifies the size of the kernel used for the filter. Larger values capture broader gradients but may lose finer details. In this code, a 5x5 kernel is used.

# How does the median filter differ from the Gaussian filter in noise reduction?
# Answer: The median filter is effective for removing salt-and-pepper noise while preserving edges. The Gaussian filter reduces noise more generally but can blur edges due to averaging based on a Gaussian distribution.

# What would happen if cv2.IMREAD_GRAYSCALE is replaced with cv2.IMREAD_COLOR in the load_img function?
# Answer: The image would be loaded in color mode (3 channels: BGR). However, the filters in the code expect a grayscale image, so it might cause errors or unintended behavior.

# Why is the result of the Sobel filter cast to np.sqrt(sobelx**2 + sobely**2)?
# Answer: To combine the gradients in the x and y directions into a single magnitude, which represents the edge strength at each pixel.

# What happens if a non-existent image path is provided as input?
# Answer: The load_img function returns None, and the program displays an error message prompting the user to provide a valid path.
```
